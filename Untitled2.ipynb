Lets see how many studets got more then 90%, more then 70, more or equal to 50% and those who failed

```python
df["total_scores"] = df["math_score"] + df["reading_score"] + df["writing_score"]
df.head()
df["avarage_score"] = df["total_scores"]/3
```

```python
def perfomance_category(score):
    if score >= 90:
        return 'Excellent'
    elif score >= 70:
        return 'Good'
    elif score >= 50:
        return 'Average'
    else:
        return 'Failed'
df["perfomance_category"]= df['avarage_score'].apply(perfomance_category)
```

```python
category_count = df['perfomance_category'].value_counts()
category_count
```

```python
category_count.plot.pie(autopct= '%1.1f%%', colors = ["gold", "lightblue", "lightgreen", "red"])
plt.title("Students Performance Category")
plt.ylabel("")
plt.show()
```
